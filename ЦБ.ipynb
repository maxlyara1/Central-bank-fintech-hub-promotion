```python
incomes = list(map(int, input().split()))
tax_rates = list(map(int, input().split()))

incomes = sorted(incomes, reverse=True)

tax_rates = sorted(tax_rates)

tax_rates = [rate / 100 for rate in tax_rates]

pairs = list(zip(incomes, tax_rates))
pairs += [(income, 0.13) for income in incomes[len(pairs):]]


result = sum(income * rate for income, rate in pairs)

print(f"{result:.4f}")
```

```python
def area_explored(A, B, N, R, r, M, C):
    total_area = A * B

    def circle_area(radius):
        return 3.14 * radius ** 2

    def triangle_area(side_length):
        return (3**(1/2) / 4) * side_length ** 2

    circle_areas = N * circle_area(R)

    triangle_areas = M * triangle_area(C)

    total_covered_area = circle_areas + triangle_areas

    explored_fraction = 1 - total_covered_area / total_area

    return round(explored_fraction, 4)

A, B, N, R, r, M, C = map(int, input().split())

result = area_explored(A, B, N, R, r, M, C)
print(result)
```
